```python
tokens = word_tokenize(document.lower())
```

```python
#buildVocab
vocab = {'<unk>':0} #initialises vocab list int form of 'token' : index number/key

Counter(tokens) #counts the number of times each token appears
```

```python
for tokens in Counter(tokens).keys():  # .keys is function which only returns unique values from all the tokens and doesnot return duplicate ones
  if tokens not in vocab:
    vocab[tokens] = len(vocab)  #len of vocab calculates the number of tokens in the vocab and assgns the new token(vocab[tokens], the next index)
vocab

```

```python
#extract sentences
input_sentences = document.split('\n') # breaks and extracts sentences from paras
```

```python
def text_to_embeddings(sentence, vocab):

  sentence_embedding = []

  for token in sentence:
    if token in vocab:
      sentence_embedding.append(vocab[token])
    else:
      sentence_embedding.append(vocab['<unk>'])
  return sentence_embedding
  #if words of the sentence present in vocab the word's index is added to sentence_embedding
  #if word not in vocab then unk token is sent whose value is 0 so 0 is added to the embedding
```

```python
input_numerical_sentences = []
for sentence in input_sentences:
 input_numerical_sentences.append(text_to_embeddings( word_tokenize(sentence.lower()), vocab))

print(input_numerical_sentences)
len(input_numerical_sentences)
```

```python
training_sequence =[]
for sentence in input_numerical_sentences:

  for i in range(1,len(sentence)):
    training_sequence.append(sentence[:i+1])

    #prints sentence by predicting next word one by one
    #starts with 10 = i, 45=am, 21 = rajiv
    #10,45
    #10,45,21
```

As we can see each trainig sequence is of different size.
we send sentences to the lstm in form of batches usually one batch is of 32 sentences . now if all sentences are of different sizes the training becomes improper so we apply padding in front as the last token becomes the target label for predicting the next word.


```python
len_list = []

for sequence in training_sequence:
  len_list.append(len(sequence))

max(len_list)
```

```python
#so we have to make each sentence of 16 size

for sequence in training_sequence:
  while len(sequence)<max(len_list):
    sequence.insert(0,0)

print(training_sequence)
```

```python
class CustomDataset(Dataset):

  def __init__(self, X, y):
    self.X = X
    self.y = y

  def __len__(self):
    return self.X.shape[0]

  def __getitem__(self, index):
    return self.X[index], self.y[index]
```

LSTM Architecture using RNN
the first word of each of the 32 sentences goes into lstm and it returns an output and hidden state which is input to another layer this continues for 15 layers and last output generated is the predicted word. the summations of all weights is taken to calculate loss and add bias to minimize loss



```python
from typing_extensions import final
class LSTMModel(nn.Module):
  def __init__(self, vocab_size):
    super().__init__()
    self.embedding = nn.Embedding(vocab_size, 100)
    self.lstm = nn.LSTM(100, 128, batch_first=True)
    self.fc = nn.Linear(128, vocab_size)

  def forward(self, x):
    embedded = self.embedding(x)
    intermediate_hidden_states, (final_hidden_state, final_cell_state) = self.lstm(embedded)
    output = self.fc(final_hidden_state.squeeze(0))
    return output
```

```python
model = LSTMModel(len(vocab))
```

```python
device = torch.device('cuda' if torch.cuda.is_available() else 'cpu')
```

```python
model.to(device)
```

```python
epochs = 50
learning_rate = 0.001

criterion = nn.CrossEntropyLoss()

optimiser = torch.optim.Adam(model.parameters(), lr=learning_rate)
```

```python
for epoch in range(epochs):
  total_loss = 0

  for batch_x,batch_y in dataloader:
    batch_x = batch_x.to(device)
    batch_y = batch_y.to(device)

    optimiser.zero_grad()

    output = model(batch_x)

    loss = criterion(output,batch_y)

    loss.backward()

    optimiser.step()

    total_loss += loss.item()

  print(f"Epoch: {epoch+1}, Loss:{total_loss:.4f}")
```
